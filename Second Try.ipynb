```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
wine_red_df = pd.read_csv("data/winequality-red.csv", sep = ";")
wine_white_df = pd.read_csv("data/winequality-white.csv", sep = ";")
```

```python
wine_red_df.describe()
```

```python
data = wine_red_df.values
```

```python
x_red = data[:,0:-1]
y_red = data[:,-1]
```

```python
x_red_train, x_red_test, y_red_train, y_red_test = train_test_split(x_red,y_red,test_size = 0.3)
```

```python
#scale = StandardScaler().fit(x_red_train)
#scale.transform(x_red_train)
#scale.transform(x_red_test)
```

```python
x_red_train,x_red_test
```

```python
scale = MinMaxScaler(feature_range = (0,10),copy = True)
x_red_train = scale.fit_transform(x_red_train)
x_red_test = scale.transform(x_red_test)
x_red_train, x_red_test
```

```python
df = pd.DataFrame(x_red_train)
df.columns = wine_red_df.columns[0:-1]
df.describe()
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
max_score = 0
opt_k = 0
for k in range(3,100):
    alg = KNeighborsClassifier(n_neighbors = k)
    alg.fit(x_red_train,y_red_train)
    score = alg.score(x_red_test,y_red_test)
    if(score>max_score):
        max_score = score
        opt_k = k
max_score, opt_k
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
max_score = 0
opt_k = 0
for k in range(3,100):
    alg = RandomForestClassifier(n_estimators = k)
    alg.fit(x_red_train,y_red_train)
    score = alg.score(x_red_test,y_red_test)
    if(score>max_score):
        max_score = score
        opt_k = k
max_score, opt_k
```
